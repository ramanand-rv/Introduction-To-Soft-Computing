## Aim: Plot activation functions

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-10, 10)
```

## Unipolar Signum

```python
def UpSignum(x):
  x1=[]
  for i in x:
    if i >= 0:
      x1.append(1)
    else:
      x1.append(0)

  return x1
```

```python
plt.plot(x, UpSignum(x))
plt.title('Unipolar Signum')
```

## Bipolar Sigmoid

```python
def BpSignum(x):
  x1=[]
  for i in x:
    if i >=0:
      x1.append(1)
    else:
      x1.append(-1)

  return x1
```

```python
plt.plot(x, BpSignum(x))
plt.title('Bipolar Signum')
```

## Unipolar Sigmoid

```python
def UpSigmoid(x):
  return (1 / (1 + np.exp(-x)))
```

```python
plt.plot(x, UpSigmoid(x))
plt.title('Unipolar Sigmoid')
```

## Bipolar Sigmoid

```python
def BpSigmoid(x):
    return ((1 - np.exp(-x)) / (1 + np.exp(-x)))
```

```python
plt.plot(x, BpSigmoid(x))
plt.title('Bipolar Sigmoid')
```
